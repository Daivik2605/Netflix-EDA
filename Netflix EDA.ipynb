### Lets see the top 10 countries from where the movies are

```python
country_counts = df['country'].value_counts().head(10) # Top 10 countries
print(country_counts)
```

```python
#create a bar chart
plt.figure(figsize = (10,6)) # Adjust figsize
bars = plt.bar(country_counts.index, country_counts.values)

#Add count values above bar
for bar, count in zip(bars, country_counts.values):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval+0.2, count, ha = 'center', va = 'bottom')

# Highlight top 3 countries
plt.bar(country_counts.index[:3], country_counts.values[:3], color = 'red')

plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 10 countries(Top 3 countries)')
plt.xticks(rotation = 45, ha = 'right')

plt.tight_layout()
plt.show()
```


bars: 
This is a list of BarContainer objects created by Matplotlib when you call a bar plotting function like plt.bar(). Each BarContainer represents an individual bar in the bar chart.

country_counts.values: 
This contains the counts (heights) of each bar. It’s typically the data that was used to create the bars.

zip(bars, country_counts.values): 
This combines the bars and country_counts.values into pairs, so you can iterate over each bar and its corresponding count together.

bar.get_height(): 
This method retrieves the height of the current bar, which corresponds to the value plotted.

bar.get_x() + bar.get_width() / 2:
This calculates the x-coordinate for placing the text. bar.get_x() gives the starting x-coordinate of the bar, 
and bar.get_width() / 2 centers the text horizontally on the bar.

yval + 0.2: 
This calculates the y-coordinate for placing the text. yval is the height of the bar, and adding 0.2 places the text slightly above the top of the bar.

count: 
This is the text to be displayed, which is the count (height) of the bar.

ha='center': 
This horizontally aligns the text to the center of the calculated x-coordinate.

va='bottom': 
This vertically aligns the text to the bottom of the calculated y-coordinate.


## Count Movies and TV Shows per country

```python
# Count Movies and TV Shows per country

movie_counts_country = df[df['type'] == 'Movie'] ['country'].value_counts()
tv_counts_country = df[df['type'] == 'TV Show'] ['country'].value_counts()
```

```python
movie_counts_country
```

```python
tv_counts_country
```

```python
# Lets combine the Movies and TV Shows counts wrt country into a single dataframe with total(use add with fill_value = 0 for 
# missing values)
df_counts = pd.DataFrame({'Movie': movie_counts_country, 'TV Show' : tv_counts_country})
df_counts
```

```python
# Now to count it wrt to country
df_counts['total_by_country'] = df_counts.sum(axis = 1)
df_counts['total_by_country']

# Sort by total count in descending order and select top 10
top_10_counts = df_counts.sort_values(by = 'total_by_country', ascending = False).head(10)
print(top_10_counts)
```

```python
top_10_counts
```

### Now lets compare the movies and TV shows for the top 10 Countries

```python
rows, cols = 2, 5
fig, axes = plt.subplots(rows, cols, figsize = (16,5)) # Adjust figsize for Visualization

# Counter to keep track of subplot position
counter = 0

# Loop through each row(country) in the dataframe
for country, row in top_10_counts.iterrows():
    # Extract the Movies, TV Shows, and Total Counts
    movie_count = row['Movie']
    tv_show_count = row['TV Show']
    total_by_count = row['total_by_country']
    
    # Create labels for pie chart slices
    labels = ['Movie', 'TV Show']
    
    # Create pie char slice sizes
    sizes = [movie_count, tv_show_count]
    
    # Select the current subplot based on counter - count
    ax = axes[counter // cols, counter % cols]
    
    # Create a pie chart from the selected subplot
    ax.pie(sizes, labels = labels, autopct = "%1.1f%%", explode = [0, 0.08], shadow = True)
    ax.set_title(country)
    
    # Increase count for the next subplot position
    counter += 1
    

plt.tight_layout()
plt.show()
```

```python
# Here we can clearly analyze which countries release more movies, like India, Egypt 
# and which countries release primarily TV Shows - South Korea, Japan
```

### Now lets analyze the ratings of the movies, tv shows

```python
# Count Movies and TV Shows per country
movie_counts_rating = df[df['type'] == 'Movie'] ['rating'].value_counts()
tv_show_counts_rating = df[df['type'] == 'TV Show'] ['rating'].value_counts()

# Combine counts into a single DataFrame with total (use add with fill_value=0 for missing values)
df_counts = pd.DataFrame({'Movie': movie_counts_rating, 'TV Show': tv_show_counts_rating})
df_counts['total_by_type'] = df_counts.sum(axis = 1)

# Sort by total count in descending order
rating_agg = df_counts.sort_values(by = 'total_by_type', ascending = False)

print(rating_agg)
# We can see the type of ratings and their total number in ascending order, also their breakdown in Movies, TV shows
```

```python
# Now lets create a dual stacked bar chart with this
fig, ax = plt.subplots(figsize = (18,9))
ax.bar(rating_agg.index, rating_agg['Movie'], label = 'Movie')
ax.bar(rating_agg.index, rating_agg['TV Show'], bottom = rating_agg['Movie'], label = 'TV Show')    

# Customize chart appearances
ax.set_xlabel('Rating', fontsize = 17)
ax.set_ylabel('Count', fontsize = 17)
ax.legend(loc = 'upper right', prop = {'size': 20})

plt.tight_layout()
plt.show()
```
